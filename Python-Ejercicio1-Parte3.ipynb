Podemos descargar datos de empresas del dataset Forbes2000 (paquete HSAUR) mediante get_rdataset(). Encontraremos un ranking de empresas categorizadas por sector, y con datos de su país, ventas, beneficios, recursos y valor de mercado.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import scipy as scipy 
import statsmodels.api as sm 
import statsmodels.formula.api as smf 
import seaborn as sns
```

```python
forbes = sm.datasets.get_rdataset('Forbes2000', 'HSAUR').data
forbes.head()
```

Seleccionar del dataframe los nombres de empresa que incluyen "elect" (sin tener en cuenta mayúsculas). Intentamos con esto obtener empresas “del sector de la electricidad” de forma aproximada.

```python
elect = forbes[forbes.name.str.contains('elect',case = False)]
```

```python
print (elect['name'].unique()) #comprobamos la selección
```

Utilizando como índice el país y la categoría y ordenándolo (sort_index), encontrar el valor de mercado de las empresas japonesas y estadounidenses.

```python
elect = elect.set_index(['country','category'])
```

```python
elect.sort_index(inplace=True)
elect.head()
```

```python
print '-----------------------------------------------'
print 'valor total electricas USA:',elect['marketvalue'].loc['United States'].sum()
print 'valor medio electricas USA:',elect['marketvalue'].loc['United States'].mean()
print '-----------------------------------------------'
print 'valor total electricas Japon:',elect['marketvalue'].loc['Japan'].sum()
print 'valor medio electricas Japon:',elect['marketvalue'].loc['Japan'].mean()
print '-----------------------------------------------'
print 'valor total electricas USA y Japon:',elect['marketvalue'].loc[['United States','Japan']].sum()
print 'valor medio electricas USA y Japon:',elect['marketvalue'].loc[['United States','Japan']].mean()
print '-----------------------------------------------'
```

Sobre el mismo subconjunto de empresas eléctricas, obtener el país donde el sector de utilities tiene mayor valor de mercado medio en sus empresas.

```python
elect_u=elect.iloc[elect.index.get_level_values('category') == 'Utilities']
elect_u
```

```python
elect_gr=elect_u['marketvalue'].groupby(level=['country']).mean()
```

```python
elect_gr
```

Observamos que el mayor valor media de utilities tiene Korea del Sur.

Volviendo al dataset original, obtener las empresas españolas.
